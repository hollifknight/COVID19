```python
import pandas as pd
countydata = pd.read_csv('/Users/holliknight/Downloads/covid-19-county-level-data.csv.zip')
```

```python
countydata.head()
```

```python
casedata = countydata.drop(columns = ['date', 'fips','deaths'])
casedata.head()
```

```python
casedata = pd.DataFrame(casedata.groupby(["state","county"]).sum())
casedata.head()
```

```python
deathdata = countydata.drop(columns = ['date', 'fips','cases'])
deathdata = pd.DataFrame(deathdata.groupby(["state","county"]).sum())
deathdata.head()
```

```python
df = pd.concat([casedata, deathdata], axis=1)
```

```python
df.head()
```

```python
df['deathrate'] = df['deaths']/df['cases']
```

```python
df.head()
```
